```python
#ONE HOT ENCODING
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder
col_trans=make_column_transformer((OneHotEncoder(handle_unknown='ignore'),['State']),remainder='passthrough')
x=col_trans.fit_transform(x)
x

```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred                  
```

```python
error=y_test-y_pred
error
```

```python
df2=pd.DataFrame({'actl':y_test,'pred':y_pred,'error':error})
df2
```

```python
from sklearn.metrics import mean_absolute_error
mbe=mean_absolute_error(y_test,y_pred)
mbe
```

```python
from sklearn.metrics import mean_absolute_percentage_error
mbpe=mean_absolute_percentage_error(y_test,y_pred)
mbpe
```

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y_test,y_pred)
mse
```

```python
RM=np.sqrt(mse)
RM
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred)
score
```
